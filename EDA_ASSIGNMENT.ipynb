```python
'''Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each.

answer:AI, ML, DL, and Data Science are related but not the same: AI is the broad goal of making
 machines “intelligent,” ML and DL are techniques used to achieve that intelligence, and Data Science
is a broader field focused on extracting insights from data (sometimes using AI/ML).

Artificial Intelligence (AI)
AI is the overall field focused on building systems that can perform tasks that normally require human intelligence, such as reasoning, planning, perception, language understanding, and decision‑making.

It includes many approaches: rule‑based systems, search and optimization, expert systems, traditional algorithms, as well as machine learning and deep learning.

Machine Learning (ML)
ML is a subset of AI that focuses on algorithms that learn patterns from data instead of being explicitly programmed with all rules.

Typical ML tasks include classification, regression, clustering, and recommendation, using methods like linear regression, decision trees, random forests, and support vector machines.

Deep Learning (DL)
DL is a subset of ML that uses multi‑layer (deep) artificial neural networks to automatically learn complex features from large amounts of data.

It is especially powerful for unstructured data such as images, audio, and text, enabling applications like image recognition, speech recognition, and large language models.

Data Science
Data Science is a broader, multidisciplinary field about collecting, cleaning, analyzing, and interpreting data to generate insights and support decisions.

A data scientist may use statistics, data visualization, database/ETL skills, and sometimes ML/DL models, but the main goal is insight and decision‑making, not necessarily building “intelligent” agents.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?

answer:Overfitting occurs when a model learns the training data too well, including noise and random fluctuations, making it perform very well on training data but poorly on unseen data because it fails to generalize. It often happens with very complex models or when training data is limited or noisy.

Underfitting happens when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and new data. This usually occurs when the model lacks complexity or is insufficiently trained.

Detection:

Overfitting is detected when there is a large gap between training accuracy (high) and test/validation accuracy (low).

Underfitting shows poor performance on both training and test data, with low accuracy and high error.

Question 3:How would you handle missing values in a dataset? Explain at least three
methods with examples.

answer:Missing values can be handled in several ways, depending on data type, amount of missingness, and how important that feature is. Here are three common methods with simple examples.

 Deletion (Dropping rows/columns)
If only a small percentage of rows have missing values, you can drop those rows so that the model trains on complete cases.
Example: In a dataset of 10,000 customers, if only 50 rows have missing “Age”, you might drop those 50 rows and keep the rest.


Simple imputation (Mean/Median/Mode)
For numerical features, replace missing values with a summary statistic like mean or median.
Example: If “Income” has values [30k, 40k, NaN, 50k], you can fill NaN with the mean (40k) or median (40k), making it [30k, 40k, 40k, 50k].

 Model-based or KNN imputation
Use other features to predict the missing value with a model or a proximity method like K-Nearest Neighbors (KNN).
Example: To fill missing “Age”, train a regression model using features like “Income”, “Education”, and “Job Type”, then predict missing ages and fill them.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

answer:An imbalanced dataset in machine learning is when the number of instances of one class (majority class) significantly outnumbers the instances of another class (minority class) in a classification problem. This imbalance causes models to be biased toward the majority class, often resulting in poor detection of the minority class, which may be critical in applications like fraud detection or medical diagnosis.​

Two techniques to handle imbalanced datasets include:

Resampling Techniques

Theoretical: These adjust the class distribution in the training data by either oversampling the minority class (adding duplicate or synthetic samples) or undersampling the majority class (removing samples) to create a more balanced dataset.

Practical: For example, SMOTE (Synthetic Minority Over-Sampling Technique) creates synthetic minority class samples from existing ones to increase its representation. Undersampling randomly removes samples from the majority to decrease its dominance.

Example: If fraud cases are only 1% in a dataset, use SMOTE to generate synthetic fraud cases or randomly remove some non-fraud cases to balance classes.

Algorithm-Level Solutions

Theoretical: Modify the learning algorithm to handle imbalance by applying higher misclassification costs to minority classes or using ensemble methods that focus on improving minority class detection.

Practical: Cost-sensitive learning changes the loss function to penalize mistakes on minority class samples more heavily. Techniques like Random Forest or XGBoost with class weights or custom objectives also improve performance on imbalanced data.

Example: Applying higher weights to fraud cases in credit card fraud detection models to reduce false negatives.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

answer:Feature scaling is important in machine learning because it ensures that numerical features of a dataset are on a similar scale, which:

Improves the performance and accuracy of many machine learning algorithms by preventing features with larger ranges from dominating others.

Speeds up convergence in gradient-based optimization algorithms like gradient descent.

Optimizes distance calculations in algorithms that rely on measures of similarity, such as k-nearest neighbors (KNN), support vector machines (SVM), and clustering algorithms.

Reduces bias towards features with larger magnitudes and enhances model interpretability by making coefficients or feature importances more comparable across features.​

min-max scaliing
scale features to fix range between 0 and 1
fixed range typically
outliers impact min and max values greatly

Standardization

Centers features at zero mean and scales to unit variance (Z-score)
No fixed range; values can be negative or positive
Outliers have less impact on mean and variance
Standardization is more robust to outliers than min-max scaling

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

answer:Label Encoding:

Assigns each unique category an integer value (e.g., "Red"=0, "Green"=1, "Blue"=2).

Creates a single numeric column.

Suitable for ordinal data where categories have a natural order (e.g., "Low," "Medium," "High").

Memory-efficient and compact.

Can mislead models if used with nominal data (no order), as it implies an ordinal relationship.

Works well with tree-based models like Decision Trees and Random Forests.

One-Hot Encoding:

Creates multiple binary columns (one per category), with a 1 indicating presence of that category and 0 otherwise.

Suitable for nominal data where no ordinal relationship exists (e.g., colors, countries).

Avoids implying any order among categories.

Increases dimensionality and memory usage, which can be costly for high-cardinality data.

Preferred for algorithms that rely on distance metrics or assume independent features, such as logistic regression, neural networks, and KNN.

When to prefer:

Use Label Encoding for ordinal categorical features or when dealing with high-cardinality categories to save memory.

Use One-Hot Encoding for nominal categorical features where no order should be assumed and when the number of categories is manageable'''



```

```python
'''Question 7: Google Play Store Dataset
a). Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?'''

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('googleplaystore.csv')

# Inspect for missing or invalid ratings and categories
df = df.dropna(subset=['Category', 'Rating'])

# Convert Ratings to numeric (handle errors)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Calculate average rating per category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Average Ratings by Category:")
print(category_ratings)

# Optionally visualize results
category_ratings.plot(kind='bar', figsize=(12,6), title='Average App Ratings by Category')
plt.ylabel('Average Rating')
plt.show()
```

```python
'''Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?'''
import pandas as pd

# Load Titanic dataset
df = pd.read_csv('titanic.csv')

# Check for missing values in 'Pclass' and 'Survived'
df = df.dropna(subset=['Pclass', 'Survived'])

# Calculate survival rate by passenger class
survival_rates = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
print("Survival Rates by Passenger Class:")
print(survival_rates)
```

```python
'''b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ≥ 18). Did children have a better chance of survival?'''
import pandas as pd

# Load Titanic dataset
df = pd.read_csv('titanic.csv')

# Drop rows with missing age or survival data
df = df.dropna(subset=['Age', 'Survived'])

# Create age groups
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

# Calculate survival rates by age group
survival_by_age = df.groupby('AgeGroup')['Survived'].mean()

print("Survival Rates by Age Group:")
print(survival_by_age)
```

```python
'''Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window'''
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv('flight_price.csv')

# a) Analysis of price variation with days left
df['days_left'] = pd.to_numeric(df['days_left'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['days_left', 'price'])

# Plot price variation over days left with possible exponential surge detection using log scale
plt.scatter(df['days_left'], df['price'], alpha=0.5)
plt.xlabel('Days to Departure')
plt.ylabel('Flight Price')
plt.title('Flight Price vs days left')
plt.yscale('log')  # Use log scale to highlight exponential surges
plt.show()
```

```python
'''Question 10: HR Analytics Dataset
a). What factors most strongly correlate with employee attrition? Use visualizations to show key
drivers (e.g., satisfaction, overtime, salary).'''
'''ans:actors strongly correlated with employee attrition in the HR Analytics dataset, typical features to consider
 include employee satisfaction levels, overtime status, salary (or pay grade), and the number of projects assigned.'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('hr_analytics.csv')  # Replace with your local file path

# Satisfaction Level distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['satisfaction_level'], bins=30, kde=True, color='blue')
plt.title('Distribution of Satisfaction Level')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.show()


# Salary distribution (assuming Salary is categorical: Low, Medium, High)
plt.figure(figsize=(6, 4))
sns.countplot(x='salary', data=df, palette='Set1')
plt.title('Distribution of Salary Levels')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()

# Boxplot of Satisfaction Level by Salary Category
plt.figure(figsize=(8, 5))
sns.boxplot(x='salary', y='satisfaction_level', data=df, palette='Set3')
plt.title('Satisfaction Level by Salary Category')
plt.xlabel('Salary')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
'''b). Are employees with more projects more likely to leave?'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('hr_analytics.csv')

# Ensure numeric conversion for projects and binary for left status
df['number_project'] = pd.to_numeric(df['number_project'], errors='coerce')
df['left'] = df['left'].astype(int)

# Calculate attrition rate by number of projects
attrition_by_projects = df.groupby('number_project')['left'].mean().reset_index()

# Plot attrition rate vs number of projects
sns.lineplot(x='number_project', y='left', data=attrition_by_projects, marker='o')
plt.title('Employee Attrition Rate vs Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Attrition Rate')
plt.show()
```
