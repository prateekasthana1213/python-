```python
'''Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

answer:Type I Error (False Positive)
A Type I error occurs when the null hypothesis is rejected when it is actually true.​

This means incorrectly concluding that there is an effect, difference, or relationship when there is not.

The probability of committing a Type I error is denoted by α
α, often referred to as the significance level.

Example: Declaring a new drug effective when it actually does nothing.

Type II Error (False Negative)
A Type II error happens when the null hypothesis is not rejected when it is actually false.​

This means failing to detect a real effect, difference, or relationship that does exist.

The probability of a Type II error is denoted by β

Example: Failing to detect that a drug is effective when it actually works.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?
*
answer:The P-value in hypothesis testing is a statistical measure that quantifies the probability of obtaining results at least as extreme as those observed,
assuming the null hypothesis is true. It helps assess the strength of evidence against the null hypothesis in the context of the test.

How to Interpret the P-value
If the P-value is less than or equal to the significance level (often 0.05), the evidence against the null hypothesis is considered strong,
 so the null hypothesis is rejected.​

If the P-value is greater than the significance level, the evidence against the null hypothesis is not strong enough, so the null hypothesis is not rejected.


Question 3:Explain the difference between a Z-test and a T-test, including when to use
each?

**************************answer:A Z-test and a T-test are both used to test hypotheses about means, but they differ mainly in the assumptions about population variance and sample size,
 which determines when each should be used.

Core Difference
Z-test uses the standard normal distribution and is appropriate when the population standard deviation σ
σ is known and/or the sample size is large (commonly n≥30).

T-test uses the Student’s t-distribution and is appropriate when the population standard deviation is unknown and the sample size is small (commonly n<30).

Question 4:What is a confidence interval, and how does the margin of error influence
its width and interpretation?

answer: confidence interval (CI) provides an estimated range within which a population parameter (like the mean or proportion) is expected to lie, based on
sample data and at a chosen confidence level (e.g., 95%).​

It reflects the uncertainty inherent in using a sample to estimate an entire population.​

Margin of Error and Its Influence
The margin of error determines the width of a confidence interval and is calculated from the sample’s standard error and a critical value
 (like a Z or t statistic corresponding to the chosen confidence level).​
 A larger margin of error results in a wider confidence interval, indicating more uncertainty in the estimate.​

A smaller margin of error narrows the interval and suggests more precision in the estimate.​


Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

answer:The purpose of an ANOVA (Analysis of Variance) test is to determine whether there are statistically significant differences between the
      means of three or more independent groups, thereby extending the basic hypothesis testing framework beyond just two groups.​

Purpose of ANOVA
ANOVA analyzes group means and variances to identify whether at least one group mean is different from the others.​

It controls the risk of Type I errors that inflate when conducting multiple t-tests instead of one overall test.​

ANOVA is widely used in research when comparing more than two treatments, groups, or experimental conditions.

Extension Beyond Two Groups
While a t-test compares the means of two groups, ANOVA tests for differences across three or more groups simultaneously.​

ANOVA examines total variance and partitions it into "between-group" and "within-group" components. It uses the F-statistic to determine
if observed differences among means are likely due to chance or represent real group differences.​

If ANOVA yields a significant result, post-hoc analyses are conducted to pinpoint which specific groups are different.'''


```

```python
'Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.'
import numpy as np
from scipy.stats import norm

# Sample dataset (Example)
data = [52, 49, 51, 50, 53, 54, 48, 50, 49, 51]

# Hypothesized population mean
population_mean = 50

# Known population standard deviation
population_std = 2  # (Assumed known)

# Calculate sample mean and size
sample_mean = np.mean(data)
n = len(data)

# Calculate Z-test statistic
z_value = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

# Significance level
alpha = 0.05

# Display results
print("Sample Mean:", sample_mean)
print("Z-Value:", z_value)
print("P-Value:", p_value)

# Interpretation of result
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis — the sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis — no significant difference from the population mean.")
```

```python
'Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using NumPy and plot the histogram.'
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000  # Number of samples

# Simulate binomial data
binomial_data = np.random.binomial(n=n, p=p, size=size)

# Plot histogram
plt.hist(binomial_data, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xticks(range(n+1))
plt.show()
```

```python
'Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.'
import numpy as np
import matplotlib.pyplot as plt

# Parameters
dist_size = 1000  # size of each sample
total_samples = 5000  # number of samples

# Generate samples from an exponential distribution (non-normal)
samples = np.random.exponential(scale=1.0, size=(total_samples, dist_size))

# Calculate the mean of each sample
sample_means = samples.mean(axis=1)

# Plot histogram of original data (one sample) and sample means
plt.figure(figsize=(12, 5))

# Histogram of data from one sample (non-normal distribution)
plt.subplot(1, 2, 1)
plt.hist(samples[0], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of One Sample (Exponential Non-Normal)')

# Histogram of sample means (CLT effect)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='orange', edgecolor='black')
plt.title('Histogram of Sample Means (CLT)')

plt.tight_layout()
plt.show()
```

```python
'Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.'
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    # Calculate sample mean and standard deviation
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Z-critical value (for normal distribution)
    z_critical = norm.ppf((1 + confidence) / 2)

    # Margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Plot visualization
    plt.figure(figsize=(6, 4))
    plt.errorbar(x=[0], y=[sample_mean],
                 yerr=[[sample_mean - lower_bound], [upper_bound - sample_mean]],
                 fmt='o', capsize=10)
    plt.title(f"{int(confidence * 100)}% Confidence Interval for Sample Mean")
    plt.ylabel("Mean")
    plt.xticks([])
    plt.grid(True)
    plt.show()

    return lower_bound, upper_bound

# Example dataset
sample_data = np.random.normal(loc=100, scale=15, size=50)
lower, upper = plot_confidence_interval(sample_data)

print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")

```

```python
'Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.'
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([25, 30, 20, 25])

# Expected frequencies (equal distribution assumption)
expected = np.array([25, 25, 25, 25])

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected distribution.")
```
